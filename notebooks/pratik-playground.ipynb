```python
import fastf1
import pandas as pd
from fastf1.ergast import Ergast

# Pandas options
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)

# Ergast options
ergast = Ergast(result_type="raw", auto_cast=True)
```

```python
event_schedule_2023 = fastf1.get_event_schedule(2023)
event_schedule_2023
```

```python
import json


event_bahraingp = event_schedule_2023.get_event_by_name("Belgian Grand Prix")

event_session = event_bahraingp.get_sprint_shootout()
event_session.load(
    laps=True,
    telemetry=False,
    weather=False,
    messages=False,
)
event_session_as_json = event_session.results.to_json(orient="records")
event_session_as_json_obj = json.loads(event_session_as_json)
event_session_as_json_obj
```

```python
race_event = event_bahraingp.get_race()
race_event.load(
    laps=True,
    telemetry=True,
    weather=True,
    messages=True,
)
race_event.results
```
